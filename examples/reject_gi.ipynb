```python
import pandas as pd
import numpy as np
import os
```

```python
# Root directory containing your folders
root_dir = '../results/reject_inference/'

dfs = []

# Walk through all directories and subdirectories
for dirpath, dirnames, filenames in os.walk(root_dir):
    for filename in filenames:
        if filename.endswith('.csv'):
            file_path = os.path.join(dirpath, filename)
            df = pd.read_csv(file_path)
            dfs.append(df.T)

# Combine all DataFrames
combined_df = pd.concat(dfs, ignore_index=False)
combined_df = combined_df.rename(columns={0: "seed", 1: "auc", 2: "ar", 3: "kickout"})
#change to int
combined_df['seed'] = combined_df['seed'].astype(int)

combined_df = combined_df.set_index('seed', append=True)
combined_df
```

```python
combined_df.groupby(level=[0]).mean()
```

```python
# Root directory containing your folders
root_dir = '../results/reject_inference_old/'

dfs = []

# Walk through all directories and subdirectories
for dirpath, dirnames, filenames in os.walk(root_dir):
    for filename in filenames:
        if filename.endswith('.csv'):
            file_path = os.path.join(dirpath, filename)
            df = pd.read_csv(file_path)
            dfs.append(df.T)

# Combine all DataFrames
old_df = pd.concat(dfs, ignore_index=False)
old_df = old_df.rename(columns={0: "seed", 1: "auc", 2: "ar", 3: "kickout"})
#change to int
old_df['seed'] = old_df['seed'].astype(int)

old_df = old_df.set_index('seed', append=True)
old_df
```

```python
mean_new = combined_df.groupby(level=[0]).mean().sort_index()
method_to_abbr = {
    'BM': 'BM',
    'aug_down': 'A-DW',
    'aug_fuzzy': 'A-FU',
    'aug_sc': 'A-SC',
    'aug_up': 'A-UW',
    'extrapolation-all': 'E-A',
    'extrapolation-confident': 'E-C',
    'extrapolation-only_1': 'E-B',
    'ls': 'LSP'
}

# Assuming df is your DataFrame with the original index
# Rename the index using the mapping
mean_new = mean_new.rename(index=method_to_abbr)
mean_new.sort_index().round(3)
```
